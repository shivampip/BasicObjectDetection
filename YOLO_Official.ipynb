### Using GPU and OpenCV

```
text= ""
with open("Makefile", 'r') as mf:
    text= mf.read().replace("GPU=0", "GPU=1")
    text= text.replace("OPENCV=0", "OPENCV=1")
with open("Makefile", "w") as f:
    f.write(text)
```

```
! cat Makefile
```

```
! make
```

```
! ./darknet detector test cfg/coco.data cfg/yolov3.cfg yolov3.weights data/kite.jpg -dont_show
```
