```python
import numpy as np
import matplotlib.pyplot as plt
from binarypipe import imagepipe as ip
```

```python
img= ip.load_img("images/55.jpg")
img= ip.img_to_array(img)
ip.plot(img)
```

```python
def rgb2gray(img):
    img= img.sum(axis= 2)
    img= img/ 3.0
    return img
```

```python
gimg= rgb2gray(img)
ip.plot(gimg)
```
